# Calculations of parametar derivatives and critical exponents

### Calculation of derivative

```python
import math
import numpy as np
import matplotlib.pyplot as plt


N = 30
x = np.linspace(0, 2*math.pi, N)
y = np.sin(x)
plt.plot(x,y)
z = (np.roll(y,1)-y)/(np.roll(x,1)-x)

plt.plot(x,z)
plt.show()
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt


N = 30
x = np.linspace(0, 2*math.pi, N)
y = x**3
plt.plot(x,y)
z = (np.roll(y,1)-y)/(np.roll(x,1)-x)

plt.plot(x,z)
plt.show()
```
